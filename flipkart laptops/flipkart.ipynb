```python
import time
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import json
import pandas as pd
from bs4 import BeautifulSoup
import requests
#from webdriver_manager.chrome import ChromeDriverManager

driver=webdriver.Chrome()
#driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get("https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=laptop%7CLaptops&requestId=60882aab-1fdd-4423-b713-f699bd08e8a4&page=27")
list_of_laptops=[]
for j in range(1,75,1):
    links1=driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']")
    links=[]
    for link in links1:
        links.append(link.get_attribute('href'))
    
    for link in links: 
        driver.execute_script("window.open();") # open a new tab
        driver.switch_to.window(driver.window_handles[1])   # switch to the tab 1
        driver.get(link)
        description=driver.find_elements(By.XPATH,"//span[@class='B_NuCI']")
        desc=[]
        for i in description:
            #print(i.text)
            desc.append(i.text)
        target=driver.find_elements(By.XPATH,"//table[@class='_14cfVK']")
        py_data = requests.get(link).text
        py_soup = BeautifulSoup (py_data, 'html.parser')
        py_table= py_soup.find ('table', attrs = {'class':'_14cfVK'})
        try:
            py_rows = py_table.find_all ('tr')
        except:
            py_rows = None
        specification=[]
        if py_rows:
            for row in py_rows:
                cols = row.find_all ('td')
                cols = [ele.text.strip() for ele in cols]
                cols = [ele for ele in cols if ele] # Get rid of empty values
                specification.append (cols)
        #print(specification)
        
            #print (cols)
        ratings=driver.find_elements(By.XPATH,"//div[@class='_3LWZlK']")
        ratings1=[]
        for i in ratings:
            #print(i.text)
            ratings1.append(i.text)
        price=driver.find_elements(By.XPATH,"//div[@class='_30jeq3 _16Jk6d']")
        price1=[]
        for i in price:
            #print(i.text)
            price1.append(i.text)
        original_price=driver.find_elements(By.XPATH,"//div[@class='_3I9_wc _2p6lqe']")
        original_price1=[]
        for i in original_price:
            #print(i.text)
            original_price1.append(i.text)
        discount=driver.find_elements(By.XPATH,"//div[@class='_3Ay6Sb _31Dcoz']")
        discount1=[]
        for i in discount:
            #print(i.text)
            discount1.append(i.text)
       # print(discount1)
        reviews=driver.find_elements(By.XPATH,"//span[@class='_2_R_DZ']")
        reviews1=[]
        for i in reviews:
            #print(i.text)
            reviews1.append(i.text)
        
        data={"Description":desc,"Specification":str(specification),"Ratings":str(ratings1),"Price":str(price1),"Original_price":str(original_price1),"Discount":str(discount1),"Reviews":str(reviews1)}

        #print(reviews1)
        #print(original_price1)
        #print(price1)
        
        #print ('Classes of each table:')
        #for table in py_soup.find_all ('table'):
        #    print (table.text)
        driver.close() # close tab 1
        driver.switch_to.window(driver.window_handles[0])
        #df=pd.DataFrame({'Description':desc,'Specification':specification,'Ratings':ratings1,'Price':price1,'Original_price':original_price1,'Discount':discount1,'Reviews':reviews1})
        df=pd.DataFrame(data)
        #data={"Description":desc,"Specification":str(specification)}
        #data1={"Description":desc,"Specification":str(specification)}
        df.columns=['Description','Specification','Ratings','Price','Original_price','Discount','Reviews']
        
        df.to_csv('flipkart5.csv',mode='a',index=False,header=False)
        list_of_laptops.append(data)
        
                               
    
        
    driver.find_element(By.XPATH,"//span[contains(text(),'Next')]").click()
    time.sleep(3)
with open('flipkart5.json', 'a') as f:
    f.write(json.dumps(list_of_laptops))


  
# adding header 

       # close tab 1
```

```python
import csv
def make_json(csvFilePath, jsonFilePath):
     
    # create a dictionary
    data = {}
     
    # Open a csv reader called DictReader
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
         
        # Convert each row into a dictionary 
        # and add it to data
        for rows in csvReader:
             
            # Assuming a column named 'No' to
            # be the primary key
            key = rows['Description']
            data[key] = rows
 
    # Open a json writer, and use the json.dumps() 
    # function to dump data
    with open(jsonFilePath, 'a', encoding='utf-8') as jsonf:
        jsonf.write(json.dumps(data, indent=4))
         
# Driver Code
 
# Decide the two file paths according to your 
# computer system
csvFilePath = r'flipkart5.csv'
jsonFilePath = r'flipkart6.json'
# Call the make_json function
make_json(csvFilePath, jsonFilePath)
```
